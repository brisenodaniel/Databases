## Terms
- The number of tuples in a table (rows) is the cardinality of the relation
- The number of attributes (columns) is the arity of the relation
## Data Type
- Atomic Types:
    - Characters: Char(20), VarChar(50)
    - Numbers: INT, BIGNINT, SMALLINT, FLOAT
    - Others: MONEY, DATETIME,...
- Every attribute must have an atomic type
    - Hence table are flat
## Table Schemas
- The __schema__ of a table is the table name, its attributes, and thier types:
    - Product (Pname: _string_, Price: _float_, Category: _string_, Manufacturer: _string_)
- A __key__ is an attribute whose values are unique; we underline a key
    - Product(Pname: string, Price: _float_, Category: _string_, Maunuacturer: _string_)
    
## Key Constraints
- A key si a minimal subset of attributes that acts as a unique identifier for tuples in a relation
- A key is an implicint constraint on which tuples can be in the relation
    - i.e. if two tuples agree on the values of the key then they must be in the same tuple!
## NULL and NOT NULL
- To say "don't know the value" we use NULL
    - NULL has (sometimes painful) semantics, more detail later
## General Constraints
- We can actually specify arbitrary assertions
    - E.G. "There cannot be 25 people in the DB class"
- In practice, we don't specify many such constraints. Why?
    - Performance reasons
- SQL Query
    - Basic form (ther eare many many more bells and whitsles)
        - SELECT attributes
        - FROM one or more relations
        - WHERE conditions
    - Call this a SFW query
- Simple SQL Query: Selection
    - Selection is the opration of filttering a reation's tuples on some condition
        - SELECT *
        - FROM Product
        - WHERE Category = 'Gadgets'
- Simple SQl Query: Projection
    - Projection is the operation of producing an output table with tuples that have a subset of thier prior attributes
        - SELECT Pname, Price, Manufacturer
        - FROM Product
        - WHERE Category = 'Gadgets'
## A Few Details
- SQL commands are case insenstive
    - same : SELECT, Select, select
- Values are not:
    - Different: 'Seattle', 'seattle'
- Use single quotes for constatns
    - 'abc' - yes
    - "abc" - no
- LIKE: Simple String Pattern Matching
    - s LIKE p: pattern matching on strings
    - p may contain two special symbls:
        - s LIKE p: pattern matching on strings
        - p may contain two special symbols:
            - % = any sequence of characters
            - _ = any singel character
    - SELECT (\*)
    - FROM Products
    - WHERE PName LIKE '%gizmo%'
- Foreign Key Constraints
    - Suppose we have the following schema:
        - Students(sid: _string_, name: _string_, gpa: _float_)
        - Enrolled(student\_id: _string_, cid: _string_, grade: _string_)
    - And we want to impose the following constraint:
        - "Only bona fide students may enroll in courses" e,e, a student must appear in the Students table to enroll in a class
    - we say that a student_id used to verify this is a __foreign key__ that refers to Students
    - What if we insert a tuple into Enrolled, but no corresponding student?
        - INSERGT is rejected (foreign keys are constraints)
        - What if we delete a student?
            1. Disallow the delete
            2. Remove all of the courses for taht student
