# What is an "Algebra"
- Mathematical sysem consisting of:
    - Operands --- Variables or values from which new values can be constructed
    - Operators --- sybmols denoting procedures that construct new values from given values
- What is Relational Algebra
    - An Algebra whose operands are relations or variables that represent relations
    - Operators are designed to do the most common thins that we need to do with relations in a database
        - The results is an algebra that can be used as a _query language_ for relations.
- Core Relational Algebra
    - Selection: picking certain rows
    - projection: picking certain columns
    - Products and joins: copositions of relations
    - Renaming of relations and attributes
- Selection
    - $R_1 := \sigma_c(R2)$
    - $C$ is a condition (as in "if" statments) that refers to attributes of $R_2$
    - $R_1$ is all those tuples of $R_2$ that satisfy $C$
- Projection
    - $R_1 := \pi_L(R_2)$
        - $L$ is a list of attributes from the schema of $R_2$
        - $R_1$ is constructed by looking at each tuple of $R_2$, extracting the attributes on $L$, in the order specified, and creating from those components a tuple for $R_1$
        - Eliminate duplicate types, if any
- Extended Projection
    - Using the same $\pi_L$ operator, we allow the list $L$ to contain arbitrary expressions involving attributes:
        1. Arithmetic on attributes, e.g., A+B -> C
        2. Duplicate occurrences of the same attribute
- Product
    - $R_3 := R_1 \times R_2$
        - Pair each tuple $t_1$ of $R_1$ with each tuple $t_2$ of $R_2$
        - Concatenation $t_1t_2$ is a tuple of $R_3$
        - Schema of $R_3$ is the attributes of $R_1$ and then $R_2$ in order
        - But bewere attribute $A$ of he same name in $R_1$ and $R_2$: use $R_{1.A}$ and $R_{2.A}$
- Theta Join
    - $R_3$ := $R_1 \bowtie_c R_2$
        - Take the product $R_1 \times R_2$
        - Then apply $\sigma_c$ to the result
    - As for $\sigma, C$ can be any bookean alued condition
    - Historic version of this poerator allowed only $A\theta B$, where $\theta$ is $=,<,$ ext; hence the name "theta-join"
- Natural Join
    - A useful join variant (_natural_ join) connects two relations by:
        - Equating attributes of the same name, and 
        - Projecting out one copy of each pair of equated attributes
   - Denoted $R_3 := R_1 \bowtie R_2$
- Renaming
    - the $\rho$ operator gives a new schema to a relation
    - $R_1 := \rho_{R_1(A_1,\dots,A_n)}(R_2)$ makes $R_1$ be a relation with attributes $A_1,\dots,A_n$ and the same tuples as $R_2$
    - Simplified notation: $R_1(A_1,\dots,A_n):=R_2$

# Aggregation
- SQL supports several aggragation operations:
    - SUM, COUNT, MIN, MAX, AVG
    - Except COUNT, all aggregations apply to a single attribute
- Aggegation: COUNT
    - COUNT applies to duplicates, unless otherwise stated
- Grouping and Aggregation
    - Semantics of the query:
        1. Copute the FROM and WHERE clauses
        2. Group by the attributes in the GROUP BY
        3. Compute the SELECT clause: grouped attributes and aggregates
- HAVING Clause
    - WHERE for groups, contains conditions on __aggregates__
    - WHERE clauses condition on __individual tuples__ 

# The Extended Algebra
- $\delta$ = elimnate duplicates
- $\tau$ = sort tuples
- $\gamma$ = grouping and aggregation
    - Outerjoin: avoids "dangling tuples" = tuples that do not join with anything.
- Duplicate Eliminaition
    - $R_1 := \delta(R_2)$
    - $R_1$ consists of one copy of each tuple that appears in $R_2$ one or more times
- Sorting
    - $R_1:=\tau_L(R_2)$
        - $L$ is a list of some of the attributes of $R_2$
    - $R_1$ is the list of tuples of $R_2$ sorted first on the value of the first attr. on $L$, then on the second attributee of $L$, and so on
        - Break ties abritrarily
    - $\tau$ is the only operator whose result is neither a set nor a bag
- Aggregation operators are not relational algebra operators
- Grouping Operator
    - $R_1 := \gamma_L(R_2)$. $L$ is a lsit of elements that are either
        1. Indivicual (grouping) attributes
        2. AGG(A), where AGG is one of the aggregation operators and $A$ is an attribute
            - An arrow and a new attribute name renames the component
